```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generating a univariate normal distribution an array of 20 by 2 samples
np.random.seed(0)
numbers = np.random.randn(20,2)
x = numbers[:,0]
y = numbers[:,1]
plt.scatter(x, y)
```

```python
# 2 by 2 matrix with random values form a uniform distribution having values between 0 and 1
matrix = np.random.uniform(low=0,high=1, size=(2,2))
print(matrix)
x1 = matrix[:,0]
y1 = matrix[:,1]
plt.scatter(x1, y1)
```

```python
# Multiplication of two matrices: numbers*matrix
multiplication = np.dot(numbers, matrix)
plt.scatter(multiplication[:,0], multiplication[:,1])
```

```python
# Calculating the variance along the data axes
x_variance = np.var(multiplication[:,0])
y_variance = np.var(multiplication[:,1])
print(x_variance,y_variance)
```

```python
# Finding Covariance Matrix
multiplication_transpose = np.transpose(multiplication)
covariance = np.cov(multiplication_transpose)
print(covariance)
```

```python
# Eigen Values and Eigen Vectors of Covariance Matrix
eigen_value, eigen_vector = np.linalg.eig(covariance)
print(eigen_value, eigen_vector)
```

```python
# PX = Y
P = eigen_vector.T

Y = np.dot(P, multiplication.T)
Y
```

```python
new_covarianceY = np.cov(Y)
new_covarianceY
```

```python
# proportion of variance = sum of eigen value of 1 to m/sum of eigen value of 1 to m to n
proportion1 = eigen_value[0]/np.sum(eigen_value)
proportion2 = eigen_value[1]/np.sum(eigen_value)
print(proportion1, proportion2)
proportion1 + proportion2
```
