```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('voice.csv')
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler
label_encoder = LabelEncoder()
```

```python
main_data = df.drop(columns='label')
target_data = df['label']
```

```python
target_data
```

```python
encoded_data = label_encoder.fit_transform(target_data)
```

```python
target_data
```

```python
encoded_data
```

```python
target_data = pd.DataFrame({'label':encoded_data})
```

```python
target_data
```

```python
main_data
```

```python
merged_df = pd.concat([main_data, target_data], axis=1)
```

```python
merged_df
```

```python
main_data.info()
```

```python
main_data.describe()
```

```python
scalar = StandardScaler()
```

```python
scaled_data = scalar.fit_transform(main_data)
```

```python
main_data = pd.DataFrame(columns=['meanfreq','sd',	'median',	'Q25',	'Q75',	'IQR',	'skew',	'kurt',	'sp.ent',	'sfm',	'mode',	'centroid',	'meanfun',	'minfun',	'maxfun',	'meandom',	'mindom',	'maxdom',	'dfrange',	'modindx'],data=scaled_data)
```

```python
main_data
```

```python
main_data.describe()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(main_data, target_data, test_size=0.2, random_state=10)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn_model_default = KNeighborsClassifier()
```

```python
knn_model_default.fit(X_train,y_train)
```

```python
y_pred = knn_model_default.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, classification_report

cm = confusion_matrix(y_test, y_pred)
```

```python
cm_plot = ConfusionMatrixDisplay(confusion_matrix=cm,
                              display_labels=knn_model_default.classes_)
cm_plot.plot()
plt.show()
```

```python
report = classification_report(y_test, y_pred)
```

```python
print(report)
```
